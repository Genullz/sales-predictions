```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
filename = ('/content/drive/MyDrive/Data Science Bootcamp/01 Week 1: Python/sales_predictions.csv')
df = pd.read_csv(filename)
df
```

A) Histogram to view the distributions of various features in your dataset.

```python
##  histogram to view the Item_Type
sns.histplot(data=df, x='Item_Type');
plt.xticks(rotation = 90);
```

```python
sns.histplot(data=df, x='Item_MRP');
plt.xticks(rotation = 90);
```

B) Boxplot to view statistical summaries of various features in your dataset

```python
df.boxplot(column='Item_Outlet_Sales');
```

```python
sns.boxplot(data=df, x='Item_Outlet_Sales');
```

```python
For_MRP = df.loc[df['Item_Type']=='MRP','Item_Outlet_Sales'].values
For_Type = df.loc[df['Item_MRP']=='TYPE','Item_Outlet_Sales'].values
plt.boxplot([For_MRP,For_Type], labels=['MRP', 'TYPE']);
```

```python
sns.boxplot(data=df, x='Item_MRP', y='Item_Type');
```

3) Heatmap of the correlation between features

```python
df = pd.read_csv('/content/drive/MyDrive/Data Science Bootcamp/01 Week 1: Python/sales_predictions.csv')
corr = df.corr()
```

```python
sns.heatmap(corr, cmap = 'Accent');
```

Adding correlation coefficients to your heatmap by using annot = True

```python
sns.heatmap(corr, cmap = 'Accent', annot = True);
```
